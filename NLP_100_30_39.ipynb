## 30. 形態素解析結果の読み込み

形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
with open("./data/nlp_100_30_39/neko.txt.mecab") as f:
    text = f.read().split('\n')
    f.close()
result = []
for word in text:
    morpheme = word.split(',')
    if morpheme[0] == 'EOS' or len(morpheme) <= 1:
        continue
    d = {
        'surface': morpheme[0].split('\t')[0],
        'base': morpheme[6],
        'pos': morpheme[0].split('\t')[1],
        'pos1': morpheme[1]
    }
    result.append(d)
result
```

## 31. 動詞

動詞の表層形をすべて抽出せよ．

```python
[d['surface'] for d in result if d['pos'] == '動詞']
```

## 32. 動詞の原形

動詞の原形をすべて抽出せよ．

```python
[d['base'] for d in result if d['pos'] == '動詞']
```

## 33. 「AのB」

2つの名詞が「の」で連結されている名詞句を抽出せよ．

```python
[''.join([d['surface'] for d in result[i:i+3]]) for i in range(len(result)-2) if result[i+1]['surface'] == 'の' and result[i]['pos'] == '名詞' and result[i+2]['pos'] == '名詞']
```

## 34. 名詞の連接

名詞の連接（連続して出現する名詞）を最長一致で抽出せよ．

```python
ls_cont_noun = []
cont_noun = ''
for d in result[:1000]:
    if d['pos'] == '名詞':
        cont_noun += d['surface']
    elif len(cont_noun:
        ls_cont_noun.append(cont_noun)
        cont_noun = ''
ls_cont_noun
```
