## 30. 形態素解析結果の読み込み

形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```python
with open("./data/nlp_100_30_39/neko.txt.mecab") as f:
    text = f.read().split('\n')
    f.close()
result = []
for word in text:
    morpheme = word.split(',')
    if morpheme[0] == 'EOS' or len(morpheme) <= 1:
        continue
    d = {
        'surface': morpheme[0].split('\t')[0],
        'base': morpheme[6],
        'pos': morpheme[0].split('\t')[1],
        'pos1': morpheme[1]
    }
    result.append(d)
result
```

## 31. 動詞

動詞の表層形をすべて抽出せよ．

```python
[d['surface'] for d in result if d['pos'] == '動詞']
```

## 32. 動詞の原形

動詞の原形をすべて抽出せよ．

```python
[d['base'] for d in result if d['pos'] == '動詞']
```

## 33. 「AのB」

2つの名詞が「の」で連結されている名詞句を抽出せよ．

```python
[''.join([d['surface'] for d in result[i:i+3]]) for i in range(len(result)-2) if result[i+1]['surface'] == 'の' and result[i]['pos'] == '名詞' and result[i+2]['pos'] == '名詞']
```

## 34. 名詞の連接

名詞の連接（連続して出現する名詞）を最長一致で抽出せよ．

```python
ls_cont_noun = []
cont_noun = []
for d in result:
    if d['pos'] == '名詞':
        cont_noun.append(d['surface'])
    elif len(cont_noun) >= 2:
        ls_cont_noun.append(''.join(cont_noun))
        cont_noun = []
ls_cont_noun
```

## 35. 単語の出現頻度

文章中に出現する単語とその出現頻度を求め，出現頻度の高い順に並べよ．

```python
from collections import Counter
words = [d['base'] for d in result]
freq = Counter(words)
```

```python
freq.most_common()
```

## 37. 「猫」と共起頻度の高い上位10語

「猫」とよく共起する（共起頻度が高い）10語とその出現頻度をグラフ（例えば棒グラフなど）で表示せよ．

```python
def check_neko(line):
    check = False
    for word in line:
        if word['base'] == '猫':
            check = True
    return check

freq_neko = []
line = []
for word in result:
    line.append(word)
    if word['base'] == '。':
        [freq_neko.append(w['base']) for w in line if check_neko(line)]
        line = []
freq_count = Counter(freq_neko)
```

```python
freq_count_pd = pd.DataFrame(freq_count.most_common(), columns={'単語': '0', '頻度': '1'})
freq_count_pd
```

```python
seaborn.barplot(x='単語', y='頻度', data=freq_count_pd[:20])
```

## 39. Zipfの法則

単語の出現頻度順位を横軸，その出現頻度を縦軸として，両対数グラフをプロットせよ．

```python
import numpy as np
```

```python
zipf = []
for i, freq in enumerate(freq_pd['頻度']):
    zipf.append([np.log(i+1), np.log(freq)])
```

```python
zipf
```

```python
zipf_pd = pd.DataFrame(zipf, columns={'順位': '0', '頻度': '1'})
zipf_pd
```

```python
seaborn.scatterplot(x='順位', y='頻度', data=zipf_pd)
```
